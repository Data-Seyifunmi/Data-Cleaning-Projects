```python
# Read the csv file
import pandas as pd
```

```python
# df means Data Frame
df = pd.read_csv("Jamb Result.csv")
```

```python
# Check the structure of the table
# df.head will by default bring the first five of the table
df.head()
```

```python
# You can specify the number of rows you want
df.head(3)
```

```python
df.head(10)
```

```python
# df.tail give the last 5 rows of the table by default
df.tail()
```

```python
# Adding the total and average column to the table
subject = ["English", "Mathematics", "Physics", "Chemistry"]
df["Total"] = df[subject].sum(axis=1)
# axis is equal to 1 because you are adding across the rows
```

```python
df.head(3)
```

```python
df.drop("Total_Score", axis=1, inplace = True)
```

```python
df.head(3)
```

```python
df["Average"] = df["Total"] / 4
```

```python
df.head(3)
```

```python
foodstuffs = ["Rice", "Beans", "Gaari", "Yam", "Semo", "Pasta", "Bread"]
foodstuffs[0]
foodstuffs[-1]
# Slicing
foodstuffs[1:6]
```

```python
# A situation where you have more then 4 subjects and you don't want to be writing the subject one after the other
# Use this method
# using "iloc", to get the position
subject = df.iloc[:, 1:5]
df["Sum"] = subject.sum(axis=1)
```

```python
df.head(3)
```

```python
test= df.iloc[0:2, 1:5]
df["test"] = test.sum(axis=1)
df.head()
```

```python
# Df.drop is use in pandas to delete a column ,
df.drop("Sum", axis =1, inplace = True)
df.drop("test", axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Calculating the overall best student in Jamb
# iloc is use for integer and index, while loc is use for name(label) and string
highest_score_index = df["Total"].idxmax()
best_student = df.loc[highest_score_index]
print(best_student)
```

```python
# specify the columns you want to print/display
# Note that if you are trying to display more than one column you have to use double square bracket[[]]
print(best_student[["Candidate_Name", "Total"]] )
```

```python
print(f"The overall best student is {best_student["Candidate_Name"]} with the total score of {best_student["Total"]}")
```

```python
# Overall Top 10 students with the highest scores
# First method
top_students = df.sort_values("Total", ascending= False)
top_students.head(10)
```

```python
# Second method, Top 5
top_students = df.nlargest(5, "Total")
top_students
```

```python
# student that score less than 200
df[df["Total"] < 200]
```
